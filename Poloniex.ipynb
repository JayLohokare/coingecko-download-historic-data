```
import requests
import json
import csv
from google.colab import files
from datetime import timedelta, date
import pandas as pd
import datetime
```

```
url = "https://poloniex.com/public?command=returnTicker"

response = requests.get(url)
ticker_data = response.json()
ticker_data = pd.DataFrame(ticker_data).transpose().reset_index()
cur_pairs = ticker_data['index']
```

```
start_time = str(int(datetime.datetime.strptime("01.01.2018 00:00:00,00", "%d.%m.%Y %H:%M:%S,%f").timestamp()))
end_time = str(int(datetime.datetime.strptime("01.05.2018 00:00:00,00", "%d.%m.%Y %H:%M:%S,%f").timestamp()))
print(start_time)
print(end_time)
```

```
for pair in cur_pairs:
    history_data = []
   
    URL = "https://poloniex.com/public?command=returnChartData" + "&currencyPair=" + pair + "&start=" + start_time + "&end=" + end_time + "&period=300"

    response = requests.get(URL)
    print(response.content)
    crypto_data = response.json()
    
    m = len(crypto_data)
    if m == 0:
        continue
    else:
        for j in range(len(crypto_data)):
            history_data.append(crypto_data[j])
    
    history_data = pd.DataFrame(history_data)
    history_data.columns = ['Close', 'Date', 'High', 'Low', 'Open', 'QuoteVolume', 'Volumn', 'WeightedAverage']
    
    history_data.to_csv( pair + ".csv", sep=',', encoding='utf-8', mode='a')
```
